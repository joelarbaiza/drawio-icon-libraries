```python
import os
import glob

# Ruta de la carpeta
folder = r"C:\Users\WIRBI\Downloads\Fabric\icons\package\dist\svg"

# Buscar archivos .svg en la carpeta
svg_files = glob.glob(os.path.join(folder, "*.svg"))

# Palabras clave a eliminar
keywords = ["filled", "regular"]

# Recorrer los archivos y eliminar los que coincidan
for file in svg_files:
    filename = os.path.basename(file).lower()  # nombre en minúsculas
    if any(keyword in filename for keyword in keywords):
        try:
            os.remove(file)
            print(f"✅ Eliminado: {file}")
        except Exception as e:
            print(f"❌ Error eliminando {file}: {e}")

print("Proceso terminado.")
```

```python
import os
import glob

# Ruta de la carpeta
folder = r"C:\Users\WIRBI\Downloads\Fabric\icons\package\dist\svg"

# Buscar archivos .svg en la carpeta
svg_files = glob.glob(os.path.join(folder, "*.svg"))

# Palabra que debe conservarse
keep_keyword = "48"

# Recorrer los archivos y eliminar los que NO tengan "48"
for file in svg_files:
    filename = os.path.basename(file).lower()  # nombre en minúsculas
    if keep_keyword not in filename:
        try:
            os.remove(file)
            print(f"🗑️ Eliminado: {file}")
        except Exception as e:
            print(f"❌ Error eliminando {file}: {e}")
    else:
        print(f"✅ Conservado: {file}")

print("Proceso terminado.")
```
